### **A quick look into Visualizing Trends in a Time Series with Python and Pandas.**

### **With data collected from Google Trends.**

**Lets use rolling windows to isolate the general trend direction of this time series.**

```
### Import packages
import pandas as pd 
import matplotlib.pyplot as plt
```

```
### Load the data into a Pandas Dataframe
pancake_data = pd.read_csv('pancakes.csv', skiprows=2, parse_dates=['Month'], index_col=['Month']) 
print(pancake_data.head())  
```

```
### Plot the data
plt.plot(pancake_data)
```

**There is some seasonality and increasing trend.**

**Lets visualize the trend of the curve with the 'rolling' method of Pandas.**

```
### Visualize the trend of the curve
y_mean = pancake_data.rolling('365D').mean()
plt.plot(y_mean)
```

**Lets add highlights to the variation each year by adding a shade to the chart with the amplitude of the Standard Deviation.**

```
### Highlight the Standard Deviation variance
y_std = pancake_data.rolling('365D').std()
plt.plot(y_mean)
plt.fill_between(y_mean.index, (y_mean - y_std).values.T[0], (y_mean + y_std).values.T[0], alpha = .5)
```
